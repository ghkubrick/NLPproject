```python
def techno_wordcount(df):
    # Using os.listdir to create a list of all of the files in dir
    LoJobs = df['Job Description'].tolist()
    LoSkills = df['Qualification Sum'].tolist()
    LoCompanies = df['Company Name'].tolist()
    LoDates = df['Publication Start Date'].tolist()
    LoCountedTechs = []
    
    for i in range(len(LoJobs)):
        DoTechs = {}
        txt1 = word_tokenize(LoJobs[i])
        txt1 = [i.lower() for i in txt1]
        txt2 = word_tokenize(LoSkills[i])
        txt2 = [i.lower() for i in txt2]
        txt = txt1 + txt2
        for word in txt:
            if [word] in lotechs:
                DoTechs[word] = DoTechs.get(word,0) + 1
        for bi in bigrams(txt):
            k = list(bi)
            if k in lotechs:
                 DoTechs[bi] = DoTechs.get(bi,0) + 1
        for tri in trigrams(txt):
            j = list(tri)
            if j in lotechs:
                DoTechs[tri] = DoTechs.get(tri,0) + 1
        if len(DoTechs) != 0:
            LoCountedTechs.append(DoTechs)
        else:
            LoCountedTechs.append(np.nan)


    res_df = pd.DataFrame(list(zip(LoCompanies, LoDates, LoCountedTechs)),
               columns =['Company', 'Date', 'CountedTechs'])
            
    return res_df
```
